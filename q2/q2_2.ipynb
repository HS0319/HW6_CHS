```python
import csv
import matplotlib.pyplot as plt

if __name__ == "__main__":
    f = open('q2.csv', 'r', encoding='cp949')
    data = list(csv.reader(f, delimiter=','))
    
    
    dice100 = list(map(int, data[0]))
    dice1000 = list(map(int, data[2]))
    dice10000 = list(map(int, data[4]))
    dice100000 = list(map(int, data[6]))
    
    f.close()
    
    
    
    plt.rc('font', family = 'Malgun Gothic')
    plt.rcParams['axes.unicode_minus'] = False
    plt.figure(figsize=(12, 12))

    plt.subplot(2, 2, 1)
    plt.hist(dice100, bins = 50)
 
    plt.xlabel('주사위 숫자')
    plt.ylabel('나온 횟수')
    plt.yticks([0,5,10,15,20,25])
    plt.title('주사위를 100번 굴렸을 때')


    plt.subplot(2, 2, 2)
    plt.hist(dice1000, bins = 50)
    plt.xlabel('주사위 숫자')
    plt.ylabel('나온 횟수')
    plt.title('주사위를 1000번 굴렸을 때')


    plt.subplot(2, 2, 3)
    plt.hist(dice10000, bins = 50)
    plt.xlabel('주사위 숫자')
    plt.ylabel('나온 횟수')
    plt.title('주사위를 10000번 굴렸을 때')


    plt.subplot(2, 2, 4)
    plt.hist(dice100000, bins = 50)
    plt.xlabel('주사위 숫자')
    plt.ylabel('나온 횟수')
    plt.title('주사위를 100000번 굴렸을 때')


    plt.show()
```
