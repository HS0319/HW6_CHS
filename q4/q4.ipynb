```python
import csv
import matplotlib.pyplot as plt

if __name__ == "__main__":
    f = open('q4.csv', 'r', encoding='cp949')
    data = csv.reader(f, delimiter=',')
    header = next(data)
    next(data)
    dict_getin = dict()
    dict_getout = dict()
    dict_getinout = dict()
    
    for row in data:
        row[10] = int(row[10])
        row[11] = int(row[11])
        row[12] = int(row[12])
        row[13] = int(row[13])

        
        dict_getin[row[1] + row[3]] = row[10] + row[12]
        dict_getout[row[1] + row[3]] = row[11] + row[13]
        dict_getinout[row[1] + row[3]] = row[10] + row[11] + row[12] + row[13]
        
    f.close()
    
        
    # 
    dict_getin = dict(sorted(dict_getin.items(), key = lambda x : x[1], reverse = True))
    dict_getout = dict(sorted(dict_getout.items(), key = lambda x : x[1], reverse = True))
    dict_getinout = dict(sorted(dict_getinout.items(), key = lambda x : x[1], reverse = True))
    
    
    dict_getin_keys = list(dict_getin.keys())
    dict_getin_values = list(dict_getin.values())
    
    dict_getout_keys = list(dict_getout.keys())
    dict_getout_values = list(dict_getout.values())
    
    dict_getinout_keys = list(dict_getinout.keys())
    dict_getinout_values = list(dict_getinout.values())
    
    plt.figure(figsize=(20, 8))
    plt.rc('font', family = 'Malgun Gothic')
    plt.rcParams['axes.unicode_minus'] = False
    
    
    
    plt.subplot(1, 3, 1)
    plt.title("  ")
    plt.xticks(rotation=90)
    plt.xlabel(' ')
    plt.ylabel('  ')
    plt.bar(dict_getin_keys[0:29], dict_getin_values[0:29])
    
    plt.subplot(1, 3, 2)
    plt.title("  ")
    plt.xlabel(' ')
    plt.ylabel('  ')
    plt.xticks(rotation=90)
    plt.bar(dict_getout_keys[0:29], dict_getout_values[0:29])
    
    plt.subplot(1, 3, 3)
    plt.title("  ")
    plt.xlabel(' ')
    plt.ylabel('  ')
    plt.xticks(rotation=90)
    plt.bar(dict_getinout_keys[0:29], dict_getinout_values[0:29])
    
    
    
    
    
```
