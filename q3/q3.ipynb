```python
import csv
import matplotlib.pyplot as plt

if __name__ == "__main__":
    xdata = ['','']
    data_2008 = []
    data_2011 = []
    data_2014 = []
    data_2017 = []
    
    f = open('q3.csv', 'r', encoding='cp949')
    data = csv.reader(f, delimiter=',')
    header = next(data)
    
    for row in data:
        data_2008.append(int(row[3]))
        data_2008.append(int(row[4]))
        data_2011.append(int(row[15]))
        data_2011.append(int(row[16]))
        data_2014.append(int(row[27]))
        data_2014.append(int(row[28]))
        data_2017.append(int(row[39]))
        data_2017.append(int(row[40]))
    
    f.close()
    
    plt.rc('font', family = 'Malgun Gothic')
    plt.rcParams['axes.unicode_minus'] = False
    plt.figure(figsize=(12, 12))
    plt.subplot(2, 2, 1)
    plt.bar(xdata, data_2008)
    plt.title('2008')
    plt.xlabel('')
    plt.ylabel(' ')

    plt.subplot(2, 2, 2)
    plt.bar(xdata, data_2011)
    plt.title('2011')
    plt.xlabel('')
    plt.ylabel(' ')

    plt.subplot(2, 2, 3)
    plt.bar(xdata, data_2014)
    plt.title('2014')
    plt.xlabel('')
    plt.ylabel(' ')

    plt.subplot(2, 2, 4)
    plt.bar(xdata, data_2017)
    plt.title('2017')
    plt.xlabel('')
    plt.ylabel(' ')

    
```
